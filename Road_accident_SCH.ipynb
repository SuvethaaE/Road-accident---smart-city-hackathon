**Dividing into Zones (north zone,south zone,east zone,west zone,central zone,NE zone,union terr)**

```python
roadcond_df =  df_road_conditions.rename(columns = {'State/ UT':'States/UTs'})
weathercond_df =  df_weather_conditions.rename(columns = {'State/ UT':'States/UTs'}) 
```

```python
North_India = ['Jammu & Kashmir', 'Punjab', 'Himachal Pradesh', 'Haryana', 'Uttarakhand', 'Uttar Pradesh', 'Chandigarh', 'Jammu and Kashmir', 'Delhi']
East_India = ['Bihar', 'Odisha', 'Jharkhand', 'West Bengal', 'Orissa']
South_India = ['Andhra Pradesh', 'Karnataka', 'Kerala' ,'Tamil Nadu', 'Telangana']
West_India = ['Rajasthan' , 'Gujarat', 'Goa','Maharashtra','Goa']
Central_India = ['Madhya Pradesh', 'Chhattisgarh']
North_East_India = ['Assam', 'Sikkim', 'Nagaland', 'Meghalaya', 'Manipur', 'Mizoram', 'Tripura', 'Arunachal Pradesh']
Union_Terr_India = ['Andaman and Nicobar Islands', 'Dadra and Nagar Haveli', 'Puducherry', 'Andaman & Nicobar Islands', 'Dadra & Nagar Haveli', 'Daman & Diu', 'Lakshadweep', 'A & N Islands', 'D & N Haveli']
```

```python
def get_zonal_names(row):
    if row['States/UTs'].strip() in North_India:
        val = 'North Zone'
    elif row['States/UTs'].strip()  in South_India:
        val = 'South Zone'
    elif row['States/UTs'].strip()  in East_India:
        val = 'East Zone'
    elif row['States/UTs'].strip()  in West_India:
        val = 'West Zone'
    elif row['States/UTs'].strip()  in Central_India:
        val = 'Central Zone'
    elif row['States/UTs'].strip()  in North_East_India:
        val = 'NE Zone'
    elif row['States/UTs'].strip()  in Union_Terr_India:
        val = 'Union Terr'
    else:
        val = 'No Value'
    return val
```

**Droping the total column changing them into Zones**

```python
df_killed_persons.drop(df_killed_persons[df_killed_persons['States/UTs'] == 'Total'].index, inplace=True)
df_killed_persons['Zones'] = df_killed_persons.apply(get_zonal_names, axis=1)
df_injured_persons.drop(df_injured_persons[df_injured_persons['States/UTs'] == 'Total'].index, inplace=True)
df_injured_persons['Zones'] = df_injured_persons.apply(get_zonal_names, axis=1)
```

```python
roadcond_df.drop(roadcond_df[roadcond_df['States/UTs'] == 'Total'].index, inplace=True)
roadcond_df['Zones'] = roadcond_df.apply(get_zonal_names, axis=1)
weathercond_df.drop(weathercond_df[weathercond_df['States/UTs'] == 'Total'].index, inplace=True)
weathercond_df['Zones'] = weathercond_df.apply(get_zonal_names, axis=1)
```

#DATA VISUALIZATION

**Zone Wise No. of Accidents where People were Killed**

```python
df_zone=pd.pivot_table(df_killed_persons, index=['Zones'],values=[2014,2015,2016,2017],aggfunc=np.sum).reset_index()
df_zone
years = [2014,2015,2016,2017]
fig,ax = plt.subplots(1,1, figsize=(20,10))
for i,year in enumerate(years):
    sns.lineplot(x=df_zone['Zones'],y=df_zone[year], label=years[i])
    plt.ylabel('Number of People Killed in Accidents')
    plt.title('Zone wise number people were killed', fontsize=15)
```

**Zone Wise No. of Accidents where People were injured**

```python
df_zone_2 =pd.pivot_table(df_injured_persons, index=['Zones'],values=[2014, 2015, 2016, 2017],aggfunc=np.sum).reset_index()
df_zone_2

years = [2014,2015,2016,2017]
fig,ax = plt.subplots(1,1, figsize=(20,10))
for i,year in enumerate(years):
    sns.lineplot(x=df_zone_2['Zones'],y=df_zone_2[year], label=years[i])
    plt.ylabel('Number of People Injured in Accidents')
    plt.title('Zone Wise number of Accidents where People were Injured', fontsize=15)
```

```python
df_south_2 = df_injured_persons[df_injured_persons['Zones'] == 'South Zone']
final2 =pd.pivot_table(df_south_2, index=['States/UTs'],values=[2014, 2015, 2016, 2017],aggfunc=np.sum).reset_index()
final2

years = [2014,2015,2016,2017]
fig,ax = plt.subplots(1,1, figsize=(20,10))
for i,year in enumerate(years):
    sns.barplot(x=final2['States/UTs'],y=final2[year])
    plt.ylabel('# of People Injured in Road Accidents')
    plt.title('Southern State Wise No. of Road Accidents where People were Injured', fontsize=15)
```

**weather condition**

Number of peopel were injured based on weather condition

```python
sub_df = weathercond_df[weathercond_df['Zones'] == 'South Zone']
df =pd.pivot_table(sub_df, index=['Zones'],aggfunc=np.sum).reset_index()
df = df.T.reset_index()
df = df.rename(columns = {'index': 'Weather Conditions', 0: 'Total'})
df = df.drop(df.index[0])
df = df.sort_values(by = ['Total'], ascending=False).head(10)
df

fig,ax = plt.subplots(1,1, figsize=(20,10))

sns.barplot(x=df['Weather Conditions'],y=df['Total'])
plt.ylabel('# of People Injured')
plt.title('Southern Zone No. of People Injured in Road Accidents Based on Weather', fontsize=15)
plt.xticks(rotation=90)
```

**Road Conditions Number of People Killed in Road Accidents**

```python
sub_df = roadcond_df[roadcond_df['Zones'] == 'South Zone']
df =pd.pivot_table(sub_df, index=['Zones'],aggfunc=np.sum).reset_index()
df = df.T.reset_index()
df = df.rename(columns = {'index': 'Road Conditions', 0: 'Total'})
df = df.drop(df.index[0])
df = df.sort_values(by = ['Total'], ascending=False).head(10)
df

fig,ax = plt.subplots(1,1, figsize=(20,10))
sns.barplot(x=df['Road Conditions'],y=df['Total'])
plt.ylabel('# of People Killed')
plt.title('Southern Zone No. of People Killed in Road Accidents Based on Road Conditions', fontsize=15)
plt.xticks(rotation=90)
```
